###Introduction

For the first assignment, I have decided to look at the New York Yankees championship teams (yes, I am not above picking a data set that may catch the interest of my professor). **The nodes of the graph are players listed on rosters of all Yankees World Series championship teams. Two players are connected if they have played on the same team.** All data was downloaded from [Baseball Reference](http://www.baseball-reference.com/). R code used to download and manipulate the data is included in the Appendix A below. Going in I did not know what to expect, but I had a suspicion that with a large amount of data the graph may be quite cumbersome. *NetworkX* is used for graphing. 

###Library and Data Import

````python
# Import relevant libraries
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# Import data from GitHub
nodes = pd.read_csv('https://raw.githubusercontent.com/ilyakats/CUNY-DATA620/master/Private/IlyaKats/yanks_nodes.csv', 
                    header=None)
edges = pd.read_csv('https://raw.githubusercontent.com/ilyakats/CUNY-DATA620/master/Private/IlyaKats/yanks_edges.csv', 
                    header=None)

# Convert data frames to lists
nodes = list(nodes.values.flatten())
edges = [tuple(x) for x in edges.to_records(index=False)]
````

###Graphing

````python
# Initial graph
G = nx.Graph()

# Loop through all nodes to customize color and size for some nodes
color_map = []
size_map = []
for i in range(0, len(nodes)-1):
    if nodes[i] in ['Babe Ruth HOF', 'Mickey Mantle HOF', 'Thurman Munson', 'Derek Jeter']:
        color_map.append('red')
        size_map.append(600)
    elif nodes[i].endswith(' HOF'):
        color_map.append('green')
        size_map.append(200)
    else:
        color_map.append('blue')
        size_map.append(30)

# Add nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)

# Draw graph
nx.draw(G, with_labels=False, nodelist=nodes, 
        node_color=color_map, node_size=size_map, 
        edge_color = 'grey')
plt.show()
````

###Analysis

Hall of Fame players are denoted by a *green* node. Babe Ruth, Mickey Mantle, Thurman Munson,  and Derek Jeter are denoted by large *red* nodes. All other players are denoted by a small *blue* node. 

Championships from 1923 to 1962 are repsented by the largest cluster (or rather a collection of clusters gathered together into an arc). Two red nodes - for Babe Ruth and Mickey Mantle - are on opposing ends of the cluster representing success in early and later years. After 1962 there was a full refresh of the team and no player on the 1977 team has won a World Series with the Yankees before. The smallest cluster represents 1977-1978 teams. The medium, odd-shaped cluster reprents 1996-2009 teams. Interestingly, it is split into two sub-clusters connected by four playes - Derek Jeter, Mariano Rivera, Jorge Posada, and Andy Pettitte. Although I do not support the *Core Four* idea spinned by the Yankees organization, the data does support this marketing trick.

###Summary and Limitations

As suspected, the graph with 485 nodes and a lot of edges is quite difficult to fully analyze in a small, non-dynamic space. However, I believe it still carries some information and at the very least made for a fun exercise. I have also used *Gephi* to browse through the graph. The data does not differentiate between everyday contributors on a team and only occassional players. The nodes are based on player's full name and if there were two players with the same name on a Yankees roster, their data may not be accurately represented. 

###Appendix A

The following R code was used to collect and manipulate the data.

```r
library(XML)
library(RCurl)
library(dplyr)
library(tidyr)
library(stringr)

yankees_ws <- c(1923, 1927, 1928, 1932, 1936, 1937, 1938, 1939, 1941, 1943,
                1947, 1949, 1950, 1951, 1952, 1953, 1956, 1958, 1961, 1962,
                1977, 1978, 1996, 1998, 1999, 2000, 2009)

all_yanks <- data.frame(Name = character(), Year = integer())

for(i in 1:length(yankees_ws)) {
  url <- getURL(str_c("http://www.baseball-reference.com/teams/NYY/",yankees_ws[i],"-roster.shtml"))
  tmp <- readHTMLTable(url, which = 1)
  tmp <- mutate(tmp[1], Year = yankees_ws[i])
  all_yanks <- rbind(all_yanks, tmp)
}

edges <- all_yanks %>% 
  inner_join(all_yanks, all_yanks, by = "Year") %>% 
  filter(Name.x != Name.y) %>% 
  select(Player1 = Name.x, Player2 = Name.y)
edges <- unique(edges[,1:2])

for(i in 1:nrow(edges)) {
  if (as.character(edges[i,1]) < as.character(edges[i,2])) {
    edges[i,] <- select(edges[i,], Player1 = Player2, Player2 = Player1)
  }
}
edges <- unique(edges[,1:2])
nodes <- unique(all_yanks[1])

write.csv(nodes, file = "nodes.csv", row.names = FALSE)
write.csv(edges, file = "edges.csv", row.names = FALSE)
write.csv(all_yanks, file = "yankees.csv", row.names = FALSE)
```
